## Frequency Encoding

```python
cols_to_frequency = ['Min Bal','Last Transaction','Multiple Acc','Active Loan','Online Banking',
                     'Which Online Banking','Active CC','Susp Transaction','Default Loan','Satisfied',
                     'if_yes','Service Call Received','Other_Account','Transaction24']
```

```python
# Frequency Encoding
d=dict()

print("Nulls\n ",'-'*50)
for cols in cols_to_frequency:
    null_val = bank[cols].isnull().sum()
    if (null_val!=0):
        print("{} : {}".format(cols,bank[cols].isnull().sum()))
        bank[cols]=bank[cols].fillna(0)
        print("filled {} : {}".format(cols,bank[cols].isnull().sum()))
    else:
        d[cols] = bank[cols].value_counts().to_dict()
        bank[cols] = bank[cols].map(d[cols])
        print("\nEncoded {} : {}".format(cols,bank[cols].value_counts()))
```
